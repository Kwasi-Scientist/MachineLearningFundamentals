### Problem 1 [10 pts]
- Implement a function `dot_product` to calculate the dot product of two 1 by n arrays. [6 pts] 
```
a = array([1, 2, 3, 4])
b = array([2, 3, 4, 5])
dot_product(a,b) = 40
```

````python
import numpy as np
import math
import pandas as pd
````

````python
### Code Starts Here
a = np.array([1, 2, 3, 4])
b = np.array([2, 3, 4, 5])
def dot_product(a, b):
    #iter through a and b
    ans = 0
    for i in range(0,a.size):
        times= a[i] *b[i]
        ans +=times 
    return ans
        

        
````

- Rotate the coordinates by 35 degress in counter clockwise. [4 pts]

````python
### Code Starts Here
#x new = cosBx1-sinBy1
#ynew = sinBx1+cosBy1
def vectorRotate(x, deg):
    #note math.cos and math.sine work in rads!
    #convert degrees to radianns
    rads = deg*0.01745329
    xnew = (math.cos(rads)*x[0])-(math.sin(rads)*x[1])
    ynew = (math.sin(rads)*x[0])+(math.cos(rads)*x[1])
    newVec = np.array([xnew, ynew])
    return newVec
````

````python
def mat_product(A,B):
    #first check that the matricies are in form mxn nxp = mxp
    if A.shape[1] != B.shape[0]:
        return "The Number of columns in A must equal the number of rows in B"
    else:
        # C = mxp
        C = np.zeros((A.shape[0],B.shape[1]))
        #iter through rows first 
        for i in range(0,C.shape[0]):
            for j in range(0,C.shape[1]):
                C[i,j] = dot_product(A[i:i+1,:].T,B[:,j:j+1])
        return C
            
        
    
````

- Inplement a function `mat_transpose` that transposes ${\bf X}$ and use the `mat_product` function to find the matrice of product: $Z={\bf X}^T\cdot{\bf Y}$ [8 pts]

        Z=mat_product(mat_transpose(X),Y)

````python
### Code Starts Here
def mat_transpose(x):
    #we want to swap the rows and columns
    transpose = np.zeros((x.shape[1],x.shape[0]))
    for i in range(0,x.shape[0]):
        for j in range(0,x.shape[1]):
            transpose[j,i] = x[i,j]
    return transpose
    
````
